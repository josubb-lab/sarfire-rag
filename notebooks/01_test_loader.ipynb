```python
import sys
sys.path.append('../src')

from loaders import PDFLoader
```

## 1. Cargar PDFs

```python
# Inicializar loader
loader = PDFLoader("../data/raw")

# Cargar todos los PDFs
documents = loader.load_all_pdfs()
```

## 2. Estadísticas

```python
loader.print_stats(documents)
```

## 3. Inspeccionar Contenido

```python
# Ver primer documento
if documents:
    doc = documents[0]
    print(f"📄 Documento: {doc['metadata']['filename']}")
    print(f"Páginas: {doc['metadata']['num_pages']}")
    print("\n" + "="*60)
    print("Primeras 500 caracteres:")
    print("="*60)
    full_text = loader.get_full_text(doc)
    print(full_text[:500])
```

## 4. Análisis de Calidad

```python
# Analizar longitud de textos
for doc in documents:
    text_lengths = [len(page['text']) for page in doc['pages']]
    avg_length = sum(text_lengths) / len(text_lengths) if text_lengths else 0
    
    print(f"\n{doc['metadata']['filename']}:")
    print(f"  - Longitud media por página: {avg_length:.0f} caracteres")
    print(f"  - Página más corta: {min(text_lengths) if text_lengths else 0} chars")
    print(f"  - Página más larga: {max(text_lengths) if text_lengths else 0} chars")
```
